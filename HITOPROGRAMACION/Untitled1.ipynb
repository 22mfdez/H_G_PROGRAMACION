```python
import matplotlib.pyplot as plt

datos2= datos.head(20) #para sacar 20 primeras libneas
counts = datos2['genres'].value_counts().reset_index()
counts.columns = ['genres', 'Count']

# Creamos el gráfico de barras con Matplotlib
plt.barh(counts['genres'], counts['Count'])

# Añadimos etiquetas y títulos
plt.xlabel('Series')
plt.ylabel('Países')
plt.title('Numero de series por país')

# Mostramos el gráfico
plt.show()
```
